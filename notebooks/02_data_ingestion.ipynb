### Objetivo

Criar banco de dados para leitura e interpretação de resultados

```python
#import modules
import os
import pandas as pd
import datetime as dt
import time as t

import plotly.graph_objects as go
from plotly.offline import plot
from pycoingecko import CoinGeckoAPI
from datetime import datetime
```

Inicializa CoinGeckoAPI e testa conexão

```python
# Inicializa o cliente da API
cg = CoinGeckoAPI()

# Testa conexão
try:
    response = cg.ping()
    if response.get("gecko_says"):
        print(f"Conexão estabelecida com sucesso: {response["gecko_says"]}")
    else:
        print(f"Conexão feita, mas resposta inesperada: {response}")
except Exception as e:
    print("FALHA na conexão: {e}")
```

Identificar caminho da pasta "data" onde será salvo as arquivos

```python
id_list_and_names = cg.get_indexes_list()

id_listname_dataframe = pd.DataFrame(id_list_and_names)

# Manter somente id e name (se existirem)
path_notebooks = os.getcwd()
parent = os.path.dirname(path_notebooks)

# Caminho até a pasta data
data_path = os.path.join(parent, "data")


# Salvar corretamente dentro de /data
id_listname_dataframe.to_csv(
    os.path.join(data_path, "id_listname_dataframe.csv"),
    index=False
)
```

Salvar IDs de criptomoedas

```python
# Salvar dentro de /data
id_listname_dataframe.to_csv(
    os.path.join(data_path, "id_listname_dataframe.csv"),
    index=False
)
```

Criar função para abertura de arquivos

```python
def open_to_csv(df_name: str = "id_listname_dataframe.csv") -> pd.DataFrame:
    """
    Lê um arquivo CSV localizado na pasta ``data`` e retorna um DataFrame do pandas.

    :param df_name: Nome do arquivo CSV a ser lido.
    :type df_name: str
    :return: DataFrame contendo os dados do arquivo CSV.
    :rtype: pandas.DataFrame
    """
    
    file_path = os.path.join(data_path, df_name)
    # Open the file using "pandas"
    dataframe = pd.read_csv(file_path)
    return dataframe

```

Buscar dados histórico do bitcoin.<br>
Obsevação: para API publica da CoinGecko é possível buscar dados em um intervalo de 365 até a data atual

```python
# Escolha da moeda
coin_id = "bitcoin"

# Converter datas para timestamp UNIX
start_date = int((datetime.now() - dt.timedelta(days=365)).timestamp())  # data atual menos 365 dias
end_date = int(t.time())                                                 # agora

# Buscar os dados
data_price_bitcoin= cg.get_coin_market_chart_range_by_id(
    id=coin_id,
    vs_currency="usd",
    from_timestamp=start_date,
    to_timestamp=end_date
)

# Converter a série de preços em DataFrame
df_price_bitcoin_daily = pd.DataFrame(data_price_bitcoin["prices"], columns=["timestamp", "price"])

# Converter timestamp para datetime
df_price_bitcoin_daily["date"] = pd.to_datetime(df_price_bitcoin_daily["timestamp"], unit="ms")

```

Salvar histórico de bitcoin em csv

```python
# Salvar corretamente dentro de /data
df_price_bitcoin_daily.to_csv(
    os.path.join(data_path, "df_price_bitcoin_daily.csv"),
    index=False
)

pd.read_csv(os.path.join(data_path,"df_price_bitcoin_daily.csv"))
```

Fazer um looping para criar os arquivos para todas as moedas existentes

```python
df_id_list = open_to_csv("id_listname_dataframe.csv")
df_id_list = df_id_list["name"]


for item in df_id_list:
    
    name = str(item.replace(" ", "_")+".csv")

    print(name)


    df_price_bitcoin_daily.to_csv(
        os.path.join(data_path, name),
        index=False
    )


```
